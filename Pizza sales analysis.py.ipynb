```python
# Caluculating selling price
df['selling_price']=df['unit_price']*df['quantity']
#Estimate cost price (assume cp is 75% of sp)
df['cost price']=df['selling_price']*0.75
#Caluculate profit
df['profit']=df['selling_price']-df['cost price']
#caluculate profit margin
df['profit margin(%)'] = (df['profit']/df['selling_price'])*100
print(df)
```

__Data Analysis and Visualization__

```python
#ANALYZE SALES TRENDS
#most and least Popular Pizza Types
pizza_counts = df['pizza_type'].value_counts()
plt.figure(figsize=(4,6))
pizza_counts.plot(kind='bar',color='red',edgecolor='black')
plt.title('Most and least popular pizza type',fontsize=16)
plt.xlabel('Pizza Type',fontsize=12)
plt.ylabel('Number of Orders',fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
#most and least Popular Sizes
pizza_counts = df['pizza_size'].value_counts()
plt.figure(figsize=(5,4))
pizza_counts.plot(kind='pie',autopct='%1.1f%%',colors=['lightcoral','gold','lightblue'],startangle=140)
plt.title('most and least Popular pizza size',fontsize=16)
plt.ylabel('')
plt.tight_layout()
plt.show
```

```python
df1=df.copy()
df['order_date']=pd.to_datetime(df['order_date'])
df['order_time']=pd.to_datetime(df['order_time'],format= '%H:%M:%S').dt.time
df['Day Name']=df['order_date'].dt.day_name()
df['Month']=df['order_date'].dt.month_name()
df['Hour']= pd.to_datetime(df['order_time'],format='%H:%M:%S').dt.hour
df.update(df1)
df
```

```python
#PEAK ORDER TIME
#peak sales time of day
sales_by_day=df['Day Name'].value_counts()
plt.figure(figsize=(6,5))
colors=['#9b59b6','#e67e22','#00cec9','#e74c3c']
sales_by_day.reindex(['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']).plot(kind='bar',color=colors,edgecolor='black')
plt.title('Peak sales by day of the week',fontsize=16)
plt.xlabel('Day of the week',fontsize=12)
plt.ylabel('Number of Orders',fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
hourly_counts = df['Hour'].value_counts().sort_index()
hourly_counts_df=hourly_counts.reset_index()
hourly_counts_df.columns=['Hour','Count']
plt.figure(figsize=(7,5))
sns.lineplot(data=hourly_counts_df,x='Hour',y='Count',marker='o',color='orange')
plt.title('peak Hourse frequency',fontsize=16)
plt.xlabel('Hour of the Day',fontsize=12)
plt.ylabel('Number of orders',fontsize=12)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.xticks(range(0,24))
plt.tight_layout()
plt.show
```

```python
sales_by_day=df['Month'].value_counts()
plt.figure(figsize=(8,5))
colors=sns.color_palette("colorblind")
sales_by_day.plot(kind='barh',color=colors,edgecolor='black')
plt.title('Peak sales by day of the Month',fontsize=16)
plt.xlabel('Month',fontsize=12)
plt.ylabel('Number of Orders',fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show
```

```python
#ANALYZE PRICING TREND
sns.scatterplot(data=df,x='unit_price',y='total_price',color='blue')
plt.xlabel('unit_price')
plt.ylabel('total_price')
plt.title('Unite Price vs Total Price')
plt.tight_layout()
plt.show()
```

```python
average=df.groupby('pizza_size')['unit_price'].mean()
sns.barplot(x=average.index,y=average.values,color='#6A5ACD')
plt.title('Average Unit Price by Size')
plt.show()
```

```python
profit_margin_summary = df.groupby(['pizza_type', 'pizza_size'])['profit_margin'].mean().reset_index()
profit_margin_summary
```

```python
df['profit_margin']=(df['profit']-df['cost_price'])/df['profit']*100
for pizza_type in df['pizza_type'].unique():
    subset = df[df['pizza_type'] == pizza_type]
    plt.bar(subset['pizza_size'], subset['profit_margin'], label=pizza_type)

plt.title('Profit Margins by Pizza Type and Size')
plt.xlabel('Pizza Size')
plt.ylabel('Profit Margin (%)')
plt.legend()
plt.show()
```
